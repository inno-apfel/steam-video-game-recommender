```python
import gzip
import pandas as pd
import numpy as np
from datetime import datetime

import matplotlib.pyplot as plt

import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import jaccard_score
from sklearn.metrics import accuracy_score
```

# Data

```python
user_reviews = pd.read_csv('data\\user_reviews_clean.csv')
user_reviews['posted'] = pd.to_datetime(user_reviews['posted'])
user_reviews.head()
```

```python
user_games = pd.read_csv('data\\user_games.csv')
user_games.head()
```

```python
games = pd.read_csv('data\\steam_games_clean.csv')
games['developer'] = games['developer'].apply(lambda developer: developer if type(developer) == str else '')
games['release_date'] = pd.to_datetime(games['release_date'])
games.head()
```

```python
user_games_info = user_games.merge(games, how='inner', left_on='item_id', right_on='id')
user_games_info.head()
```

# Exploratory Data Analysis

```python
games_by_hours_played = user_games.groupby('item_name')['playtime_forever'].sum().sort_values(ascending=False).head(25)
sns.barplot(x=games_by_hours_played.index, y=games_by_hours_played.values)
_ = plt.title('Forever Playtime by Game')
_ = plt.xlabel('games')
_ = plt.ylabel('minutes played forever (hundreds of millions)')
_ = plt.xticks(rotation=90)
```

```python
games_by_recent_hours_played = user_games.groupby('item_name')['playtime_2weeks'].sum().sort_values(ascending=False).head(25)
sns.barplot(x=games_by_recent_hours_played.index, y=games_by_recent_hours_played.values)
_ = plt.title('Recent Playtime by Game')
_ = plt.xlabel('games')
_ = plt.ylabel('minutes played forever (millions)')
_ = plt.xticks(rotation=90)
```

```python
games_by_player_count = user_games.groupby('item_name')['user_id'].nunique().sort_values(ascending=False).head(25)
sns.barplot(x=games_by_player_count.index, y=games_by_player_count.values)
_ = plt.title('Player Count by Game')
_ = plt.xlabel('games')
_ = plt.ylabel('# unique players')
_ = plt.xticks(rotation=90)
```

```python
multi_player = games[['id', 'Multiplayer']]
with_multiplayer = user_games.merge(multi_player, how='inner', left_on='item_id', right_on='id')
yes_multiplayer = with_multiplayer[with_multiplayer['Multiplayer']==1].groupby('item_name')['playtime_forever'].sum().sort_values(ascending=False).head(25)
not_multiplayer = with_multiplayer[with_multiplayer['Multiplayer']==0].groupby('item_name')['playtime_forever'].sum().sort_values(ascending=False).head(25)

ax=plt.subplots()
ax=sns.barplot(x=yes_multiplayer.index, y=yes_multiplayer.values,color = 'lightskyblue')
ax=sns.barplot(x=not_multiplayer.index, y=not_multiplayer.values,color = 'salmon')
_ = plt.title('Playtime - Multiplayer vs Singleplayer')
_ = plt.xlabel('games')
_ = plt.ylabel('minutes played forever (hundreds of millions)')
_ = plt.xticks([])
```

```python
multi_player = games[['id', 'Action']]
with_multiplayer = user_games.merge(multi_player, how='inner', left_on='item_id', right_on='id')
yes_multiplayer = with_multiplayer[with_multiplayer['Action']==1].groupby('item_name')['playtime_forever'].sum().sort_values(ascending=False).head(25)
not_multiplayer = with_multiplayer[with_multiplayer['Action']==0].groupby('item_name')['playtime_forever'].sum().sort_values(ascending=False).head(25)

ax=plt.subplots()
ax=sns.barplot(x=yes_multiplayer.index, y=yes_multiplayer.values,color = 'lightskyblue')
ax=sns.barplot(x=not_multiplayer.index, y=not_multiplayer.values,color = 'salmon')
_ = plt.title('Playtime - Action vs non-Action')
_ = plt.xlabel('games')
_ = plt.ylabel('minutes played forever (hundreds of millions)')
_ = plt.xticks([])
```

```python
multi_player = games[['id', 'Indie']]
with_multiplayer = user_games.merge(multi_player, how='inner', left_on='item_id', right_on='id')
yes_multiplayer = with_multiplayer[with_multiplayer['Indie']==1].groupby('item_name')['playtime_forever'].sum().sort_values(ascending=False).head(25)
not_multiplayer = with_multiplayer[with_multiplayer['Indie']==0].groupby('item_name')['playtime_forever'].sum().sort_values(ascending=False).head(25)

ax=plt.subplots()
ax=sns.barplot(x=not_multiplayer.index, y=not_multiplayer.values,color = 'salmon')
ax=sns.barplot(x=yes_multiplayer.index, y=yes_multiplayer.values,color = 'lightskyblue')

_ = plt.title('Playtime - Indie vs non-Indie')
_ = plt.xlabel('games')
_ = plt.ylabel('minutes played forever (hundreds of millions)')
_ = plt.xticks([])
```

```python
games_release = games[['id', 'release_date']]
with_release = user_games.merge(games_release, how='inner', left_on='item_id', right_on='id')
with_release['release_year'] = with_release['release_date'].apply(lambda date: date.year)
playtime_by_year = with_release.groupby('release_year')['playtime_forever'].sum().sort_values(ascending=False).head(25)
```

```python
sns.lineplot(x=playtime_by_year.index, y=playtime_by_year.values)
_ = plt.title('Playtime by Year')
_ = plt.xlabel('year')
_ = plt.ylabel('minutes played forever')
_ = plt.xticks(rotation=90)
```

```python
tags = ['Single-player', 'Indie', 'Action', 'Casual',
        'Adventure', 'Multi-player', 'Simulation', 'Strategy', 
        'RPG', 'Co-op', 'Cross-Platform Multiplayer']
playtime_tag = []

for tag in tags:
    playtime_tag += [user_games_info[user_games_info[tag]==1]['playtime_forever'].sum()]
```

```python
sns.barplot(x=tags, y=playtime_tag)
_ = plt.title('Playtime for Popular Tags')
_ = plt.xticks(rotation=90)
```
